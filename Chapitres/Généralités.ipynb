## Principes de base

Petit pr√©ambule pour rappeler/pr√©senter les notions n√©cessaires pour la suite. De mani√®re g√©n√©rale, en radar, on envoie un **signal √©lectromagn√©tique**, $s(t)$, d√©finit par üïÆ[1]

$$
s(t)=u(t)e^{2i\pi f_c t},
$$

compos√© du **signal en bande de base**, $u(t)$, et de **la porteuse** qui d√©cale cette composante en bande de base √† la fr√©quence porteuse $f_c$ (*carrier frequency* en anglais). Le signal en bande de base, $u(t)$, correspond √† la forme d'onde envoy√©e dans le signal radar. Tu le retrouveras souvent √©crit de la mani√®re suivante :

$$
u(t)=\frac{\textrm{Rect}\left\{\frac{t}{T}\right\}}{\sqrt{T}}e^{i\phi(t)},
$$

o√π $T$ repr√©sente la **dur√©e d'impulsion** et $\phi(t)$ **la phase**. Le terme $\textrm{Rect}\left\{\frac{t}{T}\right\}$ d√©signe la fonction porte :

$$
\textrm{Rect}\left\{\frac{t}{T}\right\}=\left \{
\begin{array}{cc}
     &  1~\textrm{for}~-\frac{T}{2}\leq t\leq \frac{T}{2}\\
     &  0~\textrm{sinon}
\end{array} \right. .
$$

On d√©finit par la suite la **fr√©quence instantan√©e** (*IF pour Instantaneous Frequency* en anglais), $f(t)$, comme la d√©riv√©e de la phase $\phi(t)$ :

$$
f(t)=\frac{1}{2\pi}\frac{d\phi(t)}{dt}.
$$

La fr√©quence instantan√©e d√©crit l'√©volution de la fr√©quence de ton signal au cours du temps.

::::{note} Remarque
Si l'√©criture $\frac{\textrm{Rect}\left\{\frac{t}{T}\right\}}{\sqrt{T}}$ peut faire peur de prime abord, c'est surtout pour les √©quations et les calculs th√©oriques. La partie $\textrm{Rect}\left\{\frac{t}{T}\right\}$ d√©finit la fonction porte comme dit plus haut et permet simplement de dire que le signal existe sur l'intervalle $[0, T]$ $\left(\left[\frac{-T}{2},\frac{T}{2}\right]\textrm{si centr√© autour de 0, comme dans la d√©finition}\right)$ et vaut $0$ ailleurs. En pratique, lorsque tu impl√©menteras des signaux, tu obtiendras ce r√©sultat sans le coder explicitement.
La partie $\frac{1}{\sqrt{T}}$ concerne juste l'amplitude du signal et permet d'avoir un signal √† puissance unitaire. Dans la suite du Notebook, en dehors des calculs th√©oriques de la fonction d'ambigu√Øt√©, on pourra consid√©rer une amplitude unitaire.
