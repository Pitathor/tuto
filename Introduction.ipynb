# Introduction

Hello toi et bienvenue,

Que tu sois stagiaire ou doctorant, si tu es ici, c'est que tu souhaites découvrir **les bases du radar et les différents outils du traitement du signal**. Ce Notebook est donc fait pour toi car il te permettra d'appréhender les notions listées ci-dessous :
- La création d'impulsions radar avec l'exemple du chirp (Représentations temporelle, fréquentielle et temps-fréquence)
- Le traitement d'un écho cible et de son retard avec le filtre adapté
- L'impact de la vitesse via la fonction d'ambiguïté
- La création de cartes distance-Doppler

Tout est implémentable via ce Notebook mais il t'est conseillé d'avoir <ins>l'environnement de programmation de ton choix</ins> (Python ou Matlab) ouvert à côté pour pouvoir bidouiller comme bon te semble (et garder une trâce sûrement utile pour toi par la suite). L'objectif principal de ce tutoriel consiste en effet à prendre en main les outils à ta manière et d'avoir un support de comparaison/vérification de ce que tu implémentes. Ce qui est implémenté dans la suite ne représente donc pas l'unique façon, loin de là, de coder les outils de traitement du signal qui deviendront tes meilleurs amis par la suite.

Bien que le Notebook soit fait pour rassembler les notions de bases qui te sont nécessaires, il ne constitue pas un recueil exhaustif. Il est donc documenté afin de te renvoyer aux ressources bibliographiques qui te permettront d'avancer. Et oui, on est un institut de recherche, faut quand même un minimum chercher pour comprendre ce que tu fais. Qui plus est, tu es libre de compléter tes connaissances comme tu le souhaites. Encore une fois, ce Notebook n'est qu'un humble support pour vérifier ce que tu comprends et implémentes. 

Bien sûr, toutes questions qui ne seraient pas répondues ici peuvent être posées à n'importe qui de l'unité. Il n'y a pas de questions cons, que des réponses qui t'attendent. 

Si tout est clair, ne perdons pas de temps et en route.
